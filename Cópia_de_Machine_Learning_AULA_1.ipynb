Standard Scaler (z-score):
$z = (x - u) / s$

Onde $x$ representa os dados, $u$ representa a média e $s$ o desvio padrão

[Saiba mais](https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.StandardScaler.html)

````python
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

padronizacao = StandardScaler()
padronizacao.fit(X)

X_padronizado = padronizacao.transform(X)
X_padronizado[0:5,0:3]
````

````python
import pandas as pd
from sklearn import datasets
from sklearn.model_selection import train_test_split

df_desb = pd.read_csv('dataset_des2.csv')
df_desb.head()
````

````python
df_desb['classe'].value_counts() #target ou rótulo ou variável classe
````

````python
# Dividindo em dados de treinamento e dados de validação (teste)

X = df_desb.drop('classe', axis=1) 
y = df_desb['classe']

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.8)
````

````python
# Número de amostras por classe nos dados de treinamento
y_train.value_counts()
````

````python
# Aplicando a técnica under sampling
from imblearn.under_sampling import RandomUnderSampler

rus = RandomUnderSampler()
X_res, y_res = rus.fit_resample(X_train, y_train)

print(y_train.shape) #treinamento
print(y_train.value_counts(), '\n')

print(y_res.shape) #nova amostragem
print(y_res.value_counts(),'\n') #quantidade de amostrar por classe
````

**Redução de dimensionalidade**

````python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets
from mpl_toolkits.mplot3d import Axes3D
````

image.png

````python
#Dividindo em X e y (Class labels)

iris = datasets.load_iris()

X = iris.data  
y = iris.target

print(X[0:5, :]) #exibindo as primeira linhas
````

O principal objetivo do PCA é reduzir o número de variáveis ​​de um conjunto de dados, preservando o máximo de informações possível.

````python
#PCA - Principal Component Analisys
from sklearn.decomposition import PCA 

#Cada componente torna-se uma combinação linear dos recursos originais
pca = PCA(n_components=3) #aplicando o PCA para obter 3 componentes
X_pca = pca.fit_transform(X)
````

````python
#Visualizando os dados

fig = plt.figure(1, figsize=(8, 6))
ax = Axes3D(fig, elev=-150, azim=110)

ax.scatter(
    X_pca[:, 0],
    X_pca[:, 1],
    X_pca[:, 2],
  
    c=y,
    cmap=plt.cm.Set1,
    edgecolor="k",
    s=40,
)
ax.set_title("PCA 3 Componentes")
ax.set_xlabel("1 Componente")
ax.w_xaxis.set_ticklabels([])
ax.set_ylabel("2 Componente")
ax.w_yaxis.set_ticklabels([])
ax.set_zlabel("3 Componente")
ax.w_zaxis.set_ticklabels([])

plt.show()
````

**Seleção de Atributos (features)**

Sklearn - Modelo.Feature_importances_

```
Retorna um array onde cada elemento dele é uma feature do seu modelo.
Quanto maior o valor, mais importante a feature é para o modelo
```

````python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

dados = load_iris()
X = dados.data     
y = dados.target 

#Criando conjunto de treino e teste
X_train, X_test, y_train, y_test = train_test_split(X, y) 

# Treinando modelo RandomForest
model = RandomForestClassifier()
model.fit(X_train, y_train) 

# Verificando a importância de cada feature para o modelo

for i in range(len(dados.feature_names)):
   print('Feature:', dados.feature_names[i], model.feature_importances_[i])
````
